```python
# -*- coding: utf-8 -*-
"""
Created on Mon Sep 21 20:06:54 2020

@author: Fausto Lorenzato

Código desenvolvido para fins didáticos.
"""

from sklearn import datasets
import numpy as np


ds = datasets.load_iris()

X = ds.data
Y = ds.target


X_0 = X[Y==0,:] # Exemplos da classe 0
X_1 = X[Y==1,:] # Exemplos da classe 1
X_2 = X[Y==2,:] # Exemplos da classe 2


X_0_mean = np.array([ np.mean(X_0[0:49,x]) for x in range(4) ]) #médias dos atributos classe 0
X_1_mean = np.array([ np.mean(X_1[0:49,x]) for x in range(4) ]) #médias dos atributos classe 1
X_2_mean = np.array([ np.mean(X_2[0:49,x]) for x in range(4) ]) #médias dos atributos classe 2

X_0_std= np.array([ np.std(X_0[0:49,x]) for x in range(4) ])  #desvio padrão dos atributos classe 0
X_1_std = np.array([ np.std(X_1[0:49,x]) for x in range(4) ]) #desvio padrão dos atributos classe 1
X_2_std = np.array([ np.std(X_2[0:49,x]) for x in range(4) ]) #desvio padrão dos atributos classe 2

def gaussian_prob(m,std,x): # vetor de médias, desvio, x
    return 1/(std*np.sqrt(2*np.pi))*np.exp(-((x-m)**2)/(2*(std**2)))





test_0 = X_0[49] #Exemplo de teste classe 0 - Exemplo não usado no treinamento
test_1 = X_1[49] #Exemplo de teste classe 1 - Exemplo não usado no treinamento
test_2 = X_2[49] #Exemplo de teste classe 2 - Exemplo não usado no treinamento

t=test_2  # Substituir aqui o exemplo de teste

p_x_0 = [gaussian_prob(X_0_mean[i],X_0_std[i],t[i]) for i in range(4)] #Verossimilhança classe 0
p_x_1 = [gaussian_prob(X_1_mean[i],X_1_std[i],t[i]) for i in range(4)] #Verossimilhança classe 1
p_x_2 = [gaussian_prob(X_2_mean[i],X_2_std[i],t[i]) for i in range(4)] #Verossimilhança classe 2




p0 = 0.33*np.prod(p_x_0)  ## 0.33 probabilidade a priori da classe 0
p1 = 0.33*np.prod(p_x_1)  ## 0.33 probabilidade a priori da classe 1
p2 = 0.33*np.prod(p_x_2)  ## 0.33 probabilidade a priori da classe 2

classe_final = np.argmax([p0,p1,p2]) # classe que maximiza

print('Classe estimada = %d'%classe_final)







```
