```python
# -*- coding: utf-8 -*-
"""
Created on Sun Sep 13 10:52:56 2020

@author: Fausto Lorenzato
"""

from matplotlib import pyplot as plt
import numpy as np

X = np.array([[0,0],[0,1],[1,0],[1,1]])

#Y = np.array( [0 , 1 , 1, 1]) # OR lógico
#Y = np.array([0,0,0,1]) # AND lógico
Y = np.array([0,1,1,0])# XOR lógico

X=(X.dot(X.T)+1)**2  # kernel

threshold = 0.6 # limiar

n = 0.01 # taxa de aprendizado

W = np.random.rand(np.shape(X)[1])
flag = 1
cont=0
while(flag==1):
    flag=0
    print('#iteration %d'%cont)
    for i in range(len(X)):
        net = X[i].dot(W) # X1*w1 + x2*w2 ... + Xn*Wn
        if net>=threshold:
            saida =1
        else:
            saida =0

        if(saida !=Y[i]):
            flag=1
            W = W + n*(Y[i] - saida)*X[i]
    cont = cont+1


plt.scatter(X[Y==0,0],X[Y==0,1])
plt.scatter(X[Y==1,0],X[Y==1,1])

X_1 = np.arange(-0.5,1.5,0.1)
X_2 = -(W[0]/W[1])*X_1 + threshold/W[1]
plt.plot(X_1,X_2)
```
