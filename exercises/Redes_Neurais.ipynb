**Código MLP**

```python
from sklearn.neural_network import MLPClassifier
import numpy as np
import pandas as pd
import random
from sklearn.metrics import accuracy_score
X = pd.read_excel('wine.xlsx')

#shuffle
lst_results =[]
for i in range(30):
  sd=i
  random.seed(sd)# rodar 30 vezes, alterando o valor do seed para cada execução
  rand_indexes = list(np.random.choice(len(X), len(X), replace=False))
  X_shuff=X.iloc[rand_indexes,:]
  #X_shuff=X
  #train,valid,test split

  train = 0.6
  valid = 0.2


  train_i = int(train*len(X))
  valid_i = int(valid*(len(X)))

  train_input = X_shuff.iloc[0:train_i,1:]
  train_target = X_shuff.iloc[0:train_i,0]

  valid_input = X_shuff.iloc[train_i:train_i+valid_i,1:]
  valid_target = X_shuff.iloc[train_i:train_i+valid_i,0]

  test_input = X_shuff.iloc[train_i+valid_i:,1:]
  test_target = X_shuff.iloc[train_i+valid_i:,0]


  #Select parameters
  #gridsearchCV
  solver='sgd' # stochastic gradient descent
  hiden_neurons = [10,20,30,40] # 4
  learning_rate = [0.1,0.01,0.001] # 3
  activation=['logistic','tanh','relu'] #3    = 4*3*3 = 36
  #[[10, 0.1, 'relu'],[10, 0.1, 'logistic']... [40, 0.001, 'tanh']]

  accuracytop = 0.0
  bestrna=0
  for h in hiden_neurons:
    for l in learning_rate:
      for a in activation:
        #[NEURON ESCONDIDOS, TAXA DE APRENDIZADO, FUNCAO DE ATIVACAO]
        rna = MLPClassifier(hidden_layer_sizes=(h,),learning_rate_init=l,activation=a,shuffle=False)
        rna.fit(train_input,train_target)

        preds = rna.predict(valid_input)
        acc=accuracy_score(preds,valid_target)

        if acc>accuracytop:
          bestrna=rna
          accuracytop=acc

  pred_test = bestrna.predict(test_input)
  acc_test = accuracy_score(pred_test,test_target)
  lst_results.append(acc_test)
print(np.mean(lst_results))
print(np.std(lst_results))
```

**Extreme Learning Machine - Regression**


```python
import numpy as np
import pandas as pd
import random
#from sklearn.neural_network import MLPRegressor
#from sklearn.metrics import mean_squared_error

data = pd.read_excel('CPU_Performance.xlsx')
data = data.iloc[:,2:]
x_input = data.iloc[:,:-2]
x_target = data.iloc[:,-2]

tr = 0.6
vl = 0.2

train_i = int(tr*len(x_input))
valid_i = int(vl*(len(x_input)))

train_input = x_input.iloc[0:train_i,:]
train_target = x_target.iloc[0:train_i]

valid_input = x_input.iloc[train_i:train_i+valid_i,:]
valid_target = x_target.iloc[train_i:train_i+valid_i]

test_input = x_input.iloc[train_i+valid_i:,:]
test_target = x_target.iloc[train_i+valid_i:]

# reescalar os dados ; minmax
(l,cols)=np.shape(train_input)
minmaxs=[]
for i in range(cols):
  min = np.min(train_input.iloc[:,i])
  max = np.max(train_input.iloc[:,i])
  minmaxs.append([min,max])

min_target = np.min(train_target)
max_target = np.max(train_target)

# (X - min)/(max - min)
for i in range(cols):
  train_input.iloc[:,i] = (train_input.iloc[:,i]-minmaxs[i][0])/(minmaxs[i][1]-minmaxs[i][0])


  valid_input.iloc[:,i] = (valid_input.iloc[:,i]-minmaxs[i][0])/(minmaxs[i][1]-minmaxs[i][0])


  test_input.iloc[:,i] = (test_input.iloc[:,i]-minmaxs[i][0])/(minmaxs[i][1]-minmaxs[i][0])


train_target = (train_target - min_target)/(max_target-min_target)
valid_target = (valid_target - min_target)/(max_target-min_target)
#test_target = (test_target - min_target)/(max_target-min_target)

(lin,cols) = np.shape(train_input)
random.seed(10)

def sigmoid(X):
  return 1/(1+np.exp(-X))


def elm(hidden_neurons,inputs,targets):
  (lin,cols) = np.shape(inputs)
  pesos = np.random.rand(cols,hidden_neurons)
  bias = np.random.rand(1,hidden_neurons)
  bias_adapt = np.ones((lin,1)).dot(bias)
  H = inputs.dot(pesos)+bias_adapt #
  #print(H)
  #print(sigmoid(H))
  H_inv = np.linalg.pinv(sigmoid(H)) # X*W=T -> W=X^{-1}*T   | RBF (Kernel gaussiano), RVFL (), autoencoder
  #treinamento rápido
  pesos_out =H_inv.dot(targets)

  return (pesos,bias,pesos_out)

def elm_predict(pesos,bias,pesos_out,inputs):
  (lin,cols) = np.shape(inputs)

  bias_adapt = np.ones((lin,1)).dot(bias)
  H = sigmoid(inputs.dot(pesos)+bias_adapt)
  preds = H.dot(pesos_out)

  return preds

hidden_neurons = [10, 20, 30, 40 ,50]
erro=10**100
for h in hidden_neurons:
  (pesos,bias,pesos_out)=elm(h,train_input,train_target)
  preds_val = elm_predict(pesos,bias,pesos_out,valid_input)

  mse_val =np.mean((preds_val - valid_target)**2)
  #print(mse_val)
  if mse_val <erro:
    best_peso,best_bias,best_pesos_out = (pesos,bias,pesos_out)
    erro = mse_val
#test

preds_test = elm_predict(best_peso,best_bias,best_pesos_out,test_input)
#print('==========================================')
#print(preds_test)
#print('==========================================')
# (X - min)/(max - min)
preds_test = preds_test*(max_target-min_target) + min_target

print('MSE test = %f'%(np.mean((test_target-preds_test)**2)))
#print(preds_test)
#print(test_target)








```
