```python

import numpy as np
from matplotlib import pyplot as plt
from sklearn import datasets

ds = datasets.load_iris()
#Largura da petala
#largura da sepala
#Comprimento da petala
#comprimento da sepala
data = ds.data
target = ds.target

meanv = np.array( [np.mean(data[:,i]) for i in range(4)])

X_prime = data- meanv

covm = X_prime.T.dot(X_prime)/(len(X_prime)-1)

eigvals, eigvetor = np.linalg.eig(covm)

K=2

P = eigvetor[:,0:K]

nova_projecao = data.dot(P)

plt.scatter(nova_projecao[target==0,0],nova_projecao[target==0,1])
plt.scatter(nova_projecao[target==1,0],nova_projecao[target==1,1])
plt.scatter(nova_projecao[target==2,0],nova_projecao[target==2,1])
plt.legend(ds.target_names)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
```
