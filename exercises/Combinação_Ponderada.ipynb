**Abordagem Estática**

```python
import pandas as pd
import numpy as np
from collections import OrderedDict
from sklearn.metrics import mean_squared_error as mse

my_dc =OrderedDict({'ARIMA':[0.1,0.2,0.3,0.23,0,1.34,2.34,4.56,0.89,0.42],'SVR':[0.18,0.23,0.37,0.28,0,1.45,2.89,3.56,1.23,1.42],'MLP':[0.17,0.56,0.23,0.78,0.56,1.78,2.23,3.34,1.89,1.23],'Target':[0.37,0.26,0.53,0.68,0.36,1.28,2.73,3.84,1.29,1.53]})
# Zhang, duas colunas : Arima, MLP || Panigrahi, duas colnunas: ETS, MLP
my_metaData = pd.DataFrame(my_dc)

X = my_metaData.iloc[:,0:3]
Y=my_metaData.iloc[:,3]

pesos = np.linalg.pinv(X).dot(Y)# Mínimos quadrados / Least squares


my_dc_teste = OrderedDict({'ARIMA':[0.58,1.2,3.3],'SVR':[1.18,0.78,2.98],'MLP':[1.17,0.56,0.76],'Target_teste':[1.0,0.9,2.0]})
my_metaData_test = pd.DataFrame(my_dc_teste)

X_teste = my_metaData_test.iloc[:,0:3].values
Y_teste = my_metaData_test.iloc[:,3].values

Previsao = X_teste.dot(pesos)

erro_estatico = mse(Y_teste,Previsao)

print(erro_estatico)


```

**Abordagem Dinâmica**

```python
import pandas as pd
import numpy as np
from collections import OrderedDict
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import mean_squared_error as mse

my_dc =OrderedDict({'ARIMA':[0.1,0.2,0.3,0.23,0,1.34,2.34,4.56,0.89,0.42],'SVR':[0.18,0.23,0.37,0.28,0,1.45,2.89,3.56,1.23,1.42],'MLP':[0.17,0.56,0.23,0.78,0.56,1.78,2.23,3.34,1.89,1.23],'Target':[0.37,0.26,0.53,0.68,0.36,1.28,2.73,3.84,1.29,1.53]})

my_metaData = pd.DataFrame(my_dc)

X = my_metaData.iloc[:,0:3]
Y=my_metaData.iloc[:,3]

my_dc_teste = OrderedDict({'ARIMA':[0.58,1.2,3.3],'SVR':[1.18,0.78,2.98],'MLP':[1.17,0.56,0.76],'Target_teste':[1.0,0.9,2.0]})
my_metaData_test = pd.DataFrame(my_dc_teste)
Y_teste = my_metaData_test.iloc[:,3].values

knn = NearestNeighbors(n_neighbors=5,metric="euclidean") # K, distancia são parâmetros importantes
knn.fit(X)
previsao_dinamica=[]
for i in range(len(my_metaData_test)):
  X_teste = my_metaData_test.iloc[i,0:3].values

  dists,indexes = knn.kneighbors(X_teste.reshape(1, -1))
  Y_kneighbors = Y[np.ravel(indexes)]
  X_kneighbors = X.iloc[np.ravel(indexes)]






  peso_dinamico = np.linalg.pinv(X_kneighbors).dot(Y_kneighbors)
  previsao_dinamica.append(X_teste.dot(peso_dinamico))



erro_dinamico = mse(Y_teste,previsao_dinamica)
print(erro_dinamico)

```
