**KNN : K-Nearest Neighbors**

```python
import pandas as pd
import numpy as np


data = pd.read_excel('realstate.xlsx')

tr_size = int(len(data) * 0.8)

X_train = data.iloc[:tr_size, 1:-1].values
y_train = data.iloc[:tr_size, -1].values

X_test = data.iloc[tr_size:, 1:-1].values
y_test = data.iloc[tr_size:, -1].values


K=5

max_values=[]
min_values=[]

X_train_N = np.zeros(X_train.shape)
X_test_N = np.zeros(X_test.shape)

for i in range(X_train[:,:-1].shape[1]):
    max_values.append(np.max(X_train[:,i]))
    min_values.append(np.min(X_train[:,i]))
    X_train_N[:,i] = (X_train[:,i] - min_values[i]) / (max_values[i] - min_values[i])
    X_test_N[:,i] = (X_test[:,i] - min_values[i]) / (max_values[i] - min_values[i])


def getNeighbors(X_test_i,X_train_N,K):
  dists=[]
  for i in range(len(X_train_N)):
    dists.append(np.linalg.norm(X_test_i-X_train_N[i,:]))
  indexes = np.argsort(dists)
  neighbors=X_train_N[indexes[0:K],:]
  return neighbors,indexes[0:K]


# rotina para teste
outputs = []
for i in range(len(X_test_N)):
  neighbors,indexes=getNeighbors(X_test_N[i,:],X_train_N,K)
  outputs.append(np.mean(y_train[indexes]))
# erro do modelo
erro = np.mean((np.array(outputs)-y_test)**2)
print("MSE: ",erro)

#KNN com técnica FIM




```

**Local Learning**

```python
import pandas as pd
import numpy as np


data = pd.read_excel('realstate.xlsx')

tr_size = int(len(data) * 0.8)

X_train = data.iloc[:tr_size, 1:-1].values
y_train = data.iloc[:tr_size, -1].values

X_test = data.iloc[tr_size:, 1:-1].values
y_test = data.iloc[tr_size:, -1].values


K=100

max_values=[]
min_values=[]

X_train_N = np.zeros(X_train.shape)
X_test_N = np.zeros(X_test.shape)

for i in range(X_train[:,:-1].shape[1]):
    max_values.append(np.max(X_train[:,i]))
    min_values.append(np.min(X_train[:,i]))
    X_train_N[:,i] = (X_train[:,i] - min_values[i]) / (max_values[i] - min_values[i])
    X_test_N[:,i] = (X_test[:,i] - min_values[i]) / (max_values[i] - min_values[i])


def getNeighbors(X_test_i,X_train_N,K):
  dists=[]
  for i in range(len(X_train_N)):
    dists.append(np.linalg.norm(X_test_i-X_train_N[i,:]))
  indexes = np.argsort(dists)
  neighbors=X_train_N[indexes[0:K],:]
  return neighbors,indexes[0:K]


# rotina para teste
outputs = []
for i in range(len(X_test_N)):
  neighbors,indexes=getNeighbors(X_test_N[i],X_train_N,K)
  X_train_neighbors = np.array(X_train_N[indexes,:]) # criando uma base de vizinhos / local
  X_train_neighbors = np.hstack((X_train_neighbors, np.ones((X_train_neighbors.shape[0], 1))))# bias/independente
  y_train_neighbors = np.array(y_train[indexes])
  coefs = np.linalg.pinv(X_train_neighbors.T.dot(X_train_neighbors)).dot(X_train_neighbors.T.dot(y_train_neighbors))
  output = np.append(X_test_N[i],1).dot(coefs)
  outputs.append(output)


# erro do modelo
erro = np.mean((np.array(outputs)-y_test)**2)
print("MSE: ",erro)
```
