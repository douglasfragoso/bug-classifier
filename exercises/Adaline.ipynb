**Código Adaline**

```python
import numpy as np
#X = [1,2,3,4,5]
#ADALINE
# 3*X+2*1 #intercept
#X**1 , X**0 -> 3*X**1 + 3*X**0
#
X_adj = np.array( [ [1,1], [2,1],[3,1],[4,1],[5,1] ])
noise = np.random.normal(0, 0.5, 5)
Y = np.array( [5,8,11,14,17] ) #+noise  #7*x + 11 [17.4337829 , 25.59115513, 32.47831352, 39.47945572, 46.44298563]
W = np.random.rand(2)
n=0.01
erro=0
somatorioErro=0
iteracoes = 1000
for j in range(iteracoes):
  for i in range(len(X_adj)):
    net = X_adj[i,:].dot(W) # x1*x1 + x2*w2 +...+xn*wn + b*1
    #funçao de ativação linear
    erro =(Y[i]-net)
    W = W +n*erro*X_adj[i,:]
    somatorioErro=somatorioErro+erro**2
  print('Erro =%f'%(somatorioErro))
  somatorioErro=0

```

**Código Mínimos Quadrados**

```python
import numpy as np

#X = [1,2,3,4,5]
# 3X+2 #intercept
#X**1 , X**0 -> 3*X**1 + 3*X**0

X_adj = np.array( [ [1,1], [2,1],[3,1],[4,1],[5,1] ])
noise = np.random.normal(0, 0.2, 5)
Y = np.array([5,8,11,14,17]  )#+noise
#linalg

W = np.linalg.pinv(X_adj).dot(Y) #    X*W =Y, X^(-1)*X*W = X^(-1)*Y = W = X^(-1)*Y

predicted=X_adj.dot(W)

mse = np.mean((Y-predicted)**2)
print(mse)
```
