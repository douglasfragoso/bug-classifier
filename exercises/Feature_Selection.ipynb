Correlação

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

raw_data = pd.read_csv('wisconsin.csv')

raw_data.head()

```

```python
data = raw_data.drop(['id'], axis=1)
data.head()

input = data.drop(['diagnosis'], axis=1)
output = data['diagnosis']


corr = input.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
input
```

```python
limiar = 0.9
selecionados = set() # evitar elementos duplicados usando set() ['perimeter_worst','perimeter_worst']

for i in range(len(corr.columns)):
    for j in range(i):
        if abs(corr.iloc[i,j]) > limiar:
            coluna_i = corr.columns[i]
            coluna_j = corr.columns[j]
            selecionados.add(coluna_i)


print(selecionados)
print(len(selecionados))
print(len(corr.columns))


input_selected = input.drop(list(selecionados), axis=1)
input_selected.head()







```

Processo de classificação usando Naive Bayes

```python
#NaiveBayes
from sklearn import naive_bayes as NB
from sklearn.model_selection import train_test_split

#Usando toda a base de dados
results_full = []
results_selected = []
for ex in range(30):

  input_train, input_test, output_train, output_test = train_test_split(input, output, test_size=0.2,random_state=ex)

  my_nb_full = NB.GaussianNB()
  my_nb_full.fit(input_train, output_train)

  results_full.append(my_nb_full.score(input_test, output_test))



  # Usando a base selecionada

  input_train_selected, input_test_selected, output_train_selected, output_test_selected = train_test_split(input_selected, output, test_size=0.2,random_state=ex)
  my_nb_selected = NB.GaussianNB()
  my_nb_selected.fit(input_train_selected, output_train_selected)

  results_selected.append(my_nb_selected.score(input_test_selected, output_test_selected))

print(np.mean(results_full))
print(np.mean(results_selected))

print(np.std(results_full))
print(np.std(results_selected))


```
